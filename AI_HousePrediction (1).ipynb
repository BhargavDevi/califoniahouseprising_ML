```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train
```

```python
y_test
```

```python
# standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)  
```

```python
x_train
```

ModelTraining

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
##print coefficient and intercept 

print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
##on which param model is trained.

regression.get_params()
```

```python
## prediction with test data


reg_pre=regression.predict(x_test)
```

```python
##plot a scatter plot for prediction


plt.scatter(y_test,reg_pre)
```

```python
reg_pre
```

## Assumptions 

```python
plt.show()
```

```python
##residules   [errors]

residules=y_test-reg_pre
```

```python
residules
```

```python
##plot residules
import seaborn as sns
sns.displot(residules,kind="kde")
```

```python
plt.show()
```

```python
## scatter plot with resp to pre
##uniform distribution
plt.scatter(reg_pre,residules)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

print("Mean Squared Error:", mean_squared_error(y_test, reg_pre))
print("Mean Absolute Error:", mean_absolute_error(y_test, reg_pre))
print(np.sqrt(mean_absolute_error(y_test, reg_pre)))
```
